```python
df = pd.read_csv('NetflixOriginals.csv')
```

```python
df['Genre'].value_counts()
```

```python
mask = df['Genre'].value_counts() > 1
df['Genre'].value_counts()[mask]
```

```python
df['Genre'].value_counts()[mask].plot(kind='bar', figsize=(10,5), 
                                      title="Most Popular Genres",
                                       xlabel='Genre', 
                                      ylabel='Number of Occurences')
                                      
```
